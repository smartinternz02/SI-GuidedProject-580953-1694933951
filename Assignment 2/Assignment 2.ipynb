```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./House_Price_India.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
numerical_features = ['number of bedrooms', 'number of bathrooms', 'living area', 'Area of the house(excluding basement)']

for feature in numerical_features:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[feature], kde=True)
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.show()
```

```python
categorical_features = ['waterfront present', 'condition of the house', 'grade of the house']

for feature in categorical_features:
    plt.figure(figsize=(8, 4))
    sns.countplot(data=df, x=feature)
    plt.title(f'Count of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()
```

```python
outlier_features = ['Price', 'living_area_renov', 'lot_area_renov']

for feature in outlier_features:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=df[feature])
    plt.title(f'Boxplot of {feature}')
    plt.xlabel(feature)
    plt.show()
```

```python
numerical_features = ['living area', 'number of bedrooms']

for feature in numerical_features:
    plt.figure(figsize=(8, 6))
    sns.scatterplot(data=df, x=feature, y='Price')
    plt.title(f'Scatter Plot of {feature} vs. Price')
    plt.xlabel(feature)
    plt.ylabel('Price')
    plt.show()
```

```python
categorical_features = ['waterfront present', 'grade of the house']

for feature in categorical_features:
    plt.figure(figsize=(8, 6))
    sns.boxplot(data=df, x=feature, y='Price')
    plt.title(f'Boxplot of {feature} vs. Price')
    plt.xlabel(feature)
    plt.ylabel('Price')
    plt.xticks(rotation=45)
    plt.show()
```

```python
numerical_features = ['living area', 'number of bedrooms', 'number of bathrooms', 'Price']

sns.pairplot(data=df[numerical_features], diag_kind='kde')
plt.suptitle('Pairplot of Numerical Variables')
plt.show()
```

```python
correlation_matrix = df[numerical_features].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='magma', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='condition of the house', hue='grade of the house')
plt.title('Relationship between Condition and Grade of the House')
plt.xlabel('Condition of the House')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Grade of the House')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df['living area'], df['number of bedrooms'], df['Price'], c='b', marker='o')
ax.set_xlabel('Living Area')
ax.set_ylabel('Number of Bedrooms')
ax.set_zlabel('Price')

plt.title('3D Scatter Plot of Living Area, Bedrooms, and Price')
plt.show()
```
